## Title: A Survey on Reinforcement Learning for Dialogue Systems

##### Link: 
[A Survey on Reinforcement Learning for Dialogue Systems](https://www.semanticscholar.org/paper/A-Survey-on-Reinforcement-Learning-for-Dialogue-Gra%C3%9Fl/cafc84a0b328f40af05fbc857621b3441316688f)


### Summary:
The paper outlines the current state-of-the-art regarding reinforcement learning methods and algorithms used in dialog systems. The authors show how RL can be used in DS, discuss current solutions, future ideas and problems, and limitations of the approach. The paper first addresses what RL is, how these principles are applied to a dialog setting, and then proposes a categorization of dialog systems in terms of domain specificity and response complexity.


### Problem/Task/Question:
The task that the paper highlights is to create an overview of current research in the area of dialog systems that work with reinforcement learning. 

### Solution/Idea/Model/Approach:
![model_image](https://d3i71xaburhd42.cloudfront.net/cafc84a0b328f40af05fbc857621b3441316688f/3-Figure3-1.png "")

* **RL in general: System learns WHAT to DO/HOW to ACT/HOW to BEHAVE**
* > Agent
* > Environment
* > set of states S = STATE SPACE
* > set of actions A = ACTION SPACE
* > Feedback Function/REWARD Function (= Function that sets the GOAL or defines what is GOOD and what is BAD)
* > VALUE function for state-action pairs (= function that evaluates each state-action pair with respect to the long-term return) Feedback Function/Reward Function (= Function that sets the GOAL or defines what is GOOD and what is BAD)
* > POLICY = function that maps states to actions, main goal of RL = develop optimal policy. 
* > exploration vs exploitation trade-off

* **Dialogue systems in general: systems with speech input and/or output**.
* Types: spoken language, written language, multi-modal systems (e.g., speech and image/video), goal-oriented or open.
* > *Input Module*: (text, speech, audio, ...) => perceive and capture text
* > *NLU (Natural Language Understanding) Module*: Understand text = understand the user's GOAL
* > *Dialogue Manager*: receives as input the understood GOAL and selects an action. Consists of WORLD MODEL/State tracker + POLICY + MEMORY(dialogue history + knowledge sources)
* > *NLG (Natural Language Generation) module* => output text (speech, chat, ...)

* **INTEGRATION of RL into DS:** ** 
* > *Feedback/Reward Function* => use positive or negative signals from the user (- i do not understand, + thank you), Mult-Modal -> sentiments, face expressions. Different feedback functions depending on the goal of the system (user satisfaction, emotional state of the user, how fast the goal was actually reached, ...).
* > Learning based on chat histories/sequences => Derive a POLICY from it
* > States = Input Queries -> CURRENT STATE, Chat History = Memory
* > Actions = Database Queries + translation of output into natural language
* > GOAL: use RL to create system responses that are context sensitive and semantically appropriate to optimally guide the user to their GOAL. 


### 3 Main Results/Findings: 
* **DS Types**: Chatbots, Infobots, Goal-oriented Bots/Task-Completion Bots, Personal Assistants.
* **Underlying DS Design Principles:** grammar/rule-based systems, probabilistic systems, neural systems.

* **To gain an overview, the authors propose the following taxonomy:**.

* a) Closed-Domain and Retrieval Based: Task is clear and domain is known. E.g. Create a flight reservation, Reserve an appointment in my calendar. Framing: Using the dialog, find out all the crucial VARIABLES (=Collection, Slot-Filling) and then call a FUNCTION and return its result in natural language. Solutions: Phrase matching, semantic parsing, statistical methods. Example: ELIZA (psychologically motivated chatbot). 

* b) Closed Domain and Generative Based: Task and domain are known, but no slot-filling is done, instead an attempt is made to find a SOLUTION and communicate it in natural language. Supervised Learning (based on example solutions) in combination with Reinforcemen Learning (to rank the goodness of the solution) is used. Example: IBM Watson: user has a specific question about a domain, such as `What is the best way to treat disease XYZ?` and based on the data, the agent computes a solution/solution path and communicates it. 

* c) Transfer Learning and Opem Domain DS: Generalize tasks across domains. Input and output are not limited to a single domain or task. E.g. Book flights, restaurants, hotels, appointments, ... => The BOT understands the concept 'BOOKING' and can apply it domain independent. Example: Microsoft Xiaoice. This domain is part of current research. 

* **The authors see the following limitations/problems:**
* > lack of semantic understanding of the bots
* > lack of robustness for diverse inputs (dialects, typing errors).
* > difficulty to define exact and appropriate reward functions (What is a positive answer? When is a response positive? How positive is an answer?)

##### **Shallow Machine LEARNING ALGORITHMS**
Shallow Machine Learning Algorithms are learning algorithms that build up a model based on pre-extracted feature representations. 

**Why do we call them shallow?** => exactly because they are not able to **learn FEATURES** on their own. 

People often say that these *shallow* methods rely on **handcrafted feature engineering**, which means that the developers/programmers themselve
have to extract the features from their data tensors and feed those into the learning algorithm. 

<img src="https://quantdare.com/wp-content/uploads/2019/06/deep_learning.png" width="500"/>

>FYI: In technical details shallow learning approaches and deep learning approaches can be unified in the framework of **computational graphs** and technically 
deep learning differentiates itself from shallow learning *only* in the characteristic of the **LOSS** function. 
Further details here: https://www.springer.com/de/book/9783319944623 

Nevertheless, there are extremely powerful (and fast and cost efficient trainable!!!) shallow methods (we will see in the practice part).

##### EVALUATION
**How can we evaluate, if our model is good or bad?** 
*Metrics* + Visualization of metrics.

* LOSS: The loss is the difference of the **target** and the **model output**.

* ACCURACY: The accuracy is the amount of **correct** model outputs in ratio to all model outputs. 
<img src="https://www.researchgate.net/publication/346510572/figure/fig2/AS:963918864318470@1606827574854/Loss-vs-Accuracy-progression-during-training.ppm" width="500"/>

* CONFUSION MATRIX: The confusion matrix tells us how our model performs, by taking into account ALL our training data and putting it into buckets which allow us to estimate the capabilities of our model. 
<img src="https://miro.medium.com/max/2102/1*fxiTNIgOyvAombPJx5KGeA.png" width="500"/>

* ROC AUC: A ROC curve plots the **true positive rate** and the **false positive rate** into one single diagram which allows us to decide if our model is adequate for the task or not (especially important in medical applications). 
AUC describes the area under the curve, ....TLDR: the greater the area, the better the model.

````python
# LEARNING

## Shallow Learning: Remember: NO feature learning, we have to find features on our own

### TRAINING
# Split your data into TRAIN, TEST, EVALUATION SET
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
#print(y_test[0])

#### Nearest Neighbor
from sklearn.neighbors import KNeighborsClassifier

# Set hyper-parameters, for controlling algorithm
nearest_neighbor_clf = KNeighborsClassifier(n_neighbors=5)

# Learn a model from training data
nearest_neighbor_clf.fit(X_train, y_train)

# Estimator state is stored in instance attributes
nearest_neighbor_clf._tree

def plot_surface(clf, X, y, 
                 xlim=(-10, 10), ylim=(-10, 10), n_steps=250, 
                 subplot=None, show=True):
    if subplot is None:
        fig = plt.figure()
    else:
        plt.subplot(*subplot)
        
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], n_steps), 
                         np.linspace(ylim[0], ylim[1], n_steps))
    
    if hasattr(clf, "decision_function"):
        z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    else:
        z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
        
    z = z.reshape(xx.shape)
    plt.contourf(xx, yy, z, alpha=0.8, cmap=plt.cm.RdBu_r)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.xlim(*xlim)
    plt.ylim(*ylim)
    
    if show:
        plt.show()
        
plot_surface(nearest_neighbor_clf, X_train, y_train)
#----------------------------------

#### Decision Trees
from sklearn.tree import DecisionTreeClassifier 
decision_tree_clf = DecisionTreeClassifier()

decision_tree_clf.fit(X_train, y_train)

def plot_clf(clf, X, y):
    plt.figure(figsize=(16, 8))
    plot_surface(clf, X, y, subplot=(1, 2, 1), show=False)
    plot_histogram(clf, X, y, subplot=(1, 2, 2), show=True)
    
def plot_histogram(clf, X, y, subplot=None, show=True):
    if subplot is None:
        fig = plt.figure()
    else:
        plt.subplot(*subplot)
    
    if hasattr(clf, "decision_function"):
        d = clf.decision_function(X)
    else:
        d = clf.predict_proba(X)[:, 1]
    
    plt.hist(d[y == "b"], bins=50, normed=True, color="b", alpha=0.5)
    plt.hist(d[y == "r"], bins=50, normed=True, color="r", alpha=0.5)
    
    if show:
        plt.show()
    
plot_surface(decision_tree_clf, X_train, y_train)

# Using even more decision trees
from sklearn.ensemble import RandomForestClassifier 
random_forest_clf = RandomForestClassifier(n_estimators=500)
random_forest_clf.fit(X_train, y_train)

plot_surface(random_forest_clf, X_train, y_train)

#### SUPPORT VECTOR MACHINE
from sklearn.svm import SVC
svm_clf = SVC(kernel="linear")  # try kernel="rbf" instead
svm_clf.fit(X_train, y_train)
#plot_surface(svm_clf, X_train, y_train)

#### NEURAL NETWORK
from sklearn.neural_network import MLPClassifier
neural_network_clf = MLPClassifier(hidden_layer_sizes=(100, 100, 100), activation="relu", learning_rate="invscaling")
neural_network_clf.fit(X_train, y_train)
plot_surface(neural_network_clf, X_train, y_train)

# Learning Curves
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit
def plot_learning_curve(estimator, title, X, y, axes=None, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate 3 plots: the test and training learning curve, the training
    samples vs fit times curve, the fit times vs score curve.

    Parameters
    ----------
    estimator : estimator instance
        An estimator instance implementing `fit` and `predict` methods which
        will be cloned for each validation.

    title : str
        Title for the chart.

    X : array-like of shape (n_samples, n_features)
        Training vector, where ``n_samples`` is the number of samples and
        ``n_features`` is the number of features.

    y : array-like of shape (n_samples) or (n_samples, n_features)
        Target relative to ``X`` for classification or regression;
        None for unsupervised learning.

    axes : array-like of shape (3,), default=None
        Axes to use for plotting the curves.

    ylim : tuple of shape (2,), default=None
        Defines minimum and maximum y-values plotted, e.g. (ymin, ymax).

    cv : int, cross-validation generator or an iterable, default=None
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:

          - None, to use the default 5-fold cross-validation,
          - integer, to specify the number of folds.
          - :term:`CV splitter`,
          - An iterable yielding (train, test) splits as arrays of indices.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : int or None, default=None
        Number of jobs to run in parallel.
        ``None`` means 1 unless in a :obj:`joblib.parallel_backend` context.
        ``-1`` means using all processors. See :term:`Glossary <n_jobs>`
        for more details.

    train_sizes : array-like of shape (n_ticks,)
        Relative or absolute numbers of training examples that will be used to
        generate the learning curve. If the ``dtype`` is float, it is regarded
        as a fraction of the maximum size of the training set (that is
        determined by the selected validation method), i.e. it has to be within
        (0, 1]. Otherwise it is interpreted as absolute sizes of the training
        sets. Note that for classification the number of samples usually have
        to be big enough to contain at least one sample from each class.
        (default: np.linspace(0.1, 1.0, 5))
    """
    if axes is None:
        _, axes = plt.subplots(1, 3, figsize=(20, 5))

    axes[0].set_title(title)
    if ylim is not None:
        axes[0].set_ylim(*ylim)
    axes[0].set_xlabel("Training examples")
    axes[0].set_ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = \
        learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                       train_sizes=train_sizes,
                       return_times=True)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Plot learning curve
    axes[0].grid()
    axes[0].fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    axes[0].fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
    axes[0].plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
    axes[0].plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
    axes[0].legend(loc="best")

    return plt

fig, axes = plt.subplots(3, 1, figsize=(10, 15))

#X, y = load_digits(return_X_y=True)

title = "Learning Curves (Neural Networks)"
# Cross validation with 10 iterations to get smoother mean test and train
# score curves, each time with 20% data randomly selected as a validation set.
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)

estimator = MLPClassifier()
plot_learning_curve(estimator, title, X, y, axes=axes[:,], ylim=(0.7, 1.01),
                    cv=cv, n_jobs=4)
plt.show()

#### ADDITIONAL HOMEWORK: 
#### Logistic Regression
#from sklearn.linear_model import LogisticRegression
#clf = LogisticRegression()
#clf.fit(X, y)
#plot_surface(clf, X_train, y_train)

#### Gaussian Processes
#from sklearn.gaussian_process import GaussianProcessClassifier
#clf = GaussianProcessClassifier()
#clf.fit(X, y)
#plot_surface(clf, X_train, y_train)
````
